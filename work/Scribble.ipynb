```python
import numpy as np
import matplotlib.pyplot as plt
from beakerx import *
from pyscribble.scribble import scribble
```

```python
def wedding(word, f, title="Lydia & Thomas, August 17, Rhodes House"):
    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8.27, 11.69), subplot_kw={"aspect": "equal"})
    
    scribble(axes[1], text=word, f=f, title=title, linewidth=2, color="blue")
               
    axes[0].set_axis_off()   
    axes[0].text(0.5, 0.5, '{name}\n{function}'.format(name=word, function=f), horizontalalignment='right', rotation=180) 
    
    fig.savefig("data/{name}.pdf".format(name=word))
```

```python
form = EasyForm("Create a name tag")
form.addTextField("Text")
form["Text"] = "Thomas Schmelzer"
form.addTextField("Title")
form["Title"] = "Lydia & Thomas, August 17, Rhodes House"
form.addTextField("Function")
form["Function"] = "exp((-1+2j)*z)"
form.addButton("Go!", tag="run")
form
```

```python
wedding(form["Text"], f=form["Function"], title=form["Title"])
```
